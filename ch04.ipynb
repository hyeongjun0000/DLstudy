```python
import pandas as pd
import folium
import json
```

```python
file_path ="/Users/kimhyeongjun/Desktop/dlstudy/경기도인구데이터.xlsx"
df = pd.read_excel(file_path, index_col='구분')
df.columns = df.columns.map(str)

geo_path = '/Users/kimhyeongjun/Desktop/dlstudy/경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding ='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding = 'utf-8-sig'))

g_map = folium.Map(location=[37.55, 126.98],zoom_start=9)

year = '2017'
```

```python
folium.Choropleth(geo_data = geo_data, data = df[year], columns=[df.index, df[year]],
        fill_color='YlOrRd',fill_opacity=0.7, line_opacity=0.3,
        threshold_scale=[10000,100000,300000,500000,700000],
        key_on ='feature.properties.name',legend_name="인구수").add_to(g_map)

g_map.save('경기도인구'+year +'.html')
```
